```python
import pandas as pd
import joblib

# Load features
df = pd.read_csv("../data/generated/tourist_features.csv")

# Load models
risk_model = joblib.load("../models/risk_model.pkl")
iso_model = joblib.load("../models/isolation_forest.pkl")
```

```python
X = df.drop(columns=["tourist_id"])

df["risk_prediction"] = risk_model.predict(X)
df["anomaly_prediction"] = iso_model.predict(X)

df["anomaly_prediction"] = df["anomaly_prediction"].apply(
    lambda x: 1 if x == -1 else 0
)
```

```python
def compute_severity(row):
    if row["risk_prediction"] == 1 and row["anomaly_prediction"] == 1:
        return "CRITICAL"
    elif row["risk_prediction"] == 1:
        return "HIGH RISK"
    elif row["anomaly_prediction"] == 1:
        return "SUSPICIOUS"
    else:
        return "SAFE"

df["final_status"] = df.apply(compute_severity, axis=1)

df[["tourist_id", "risk_prediction", "anomaly_prediction", "final_status"]].head()
```
